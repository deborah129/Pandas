```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(20)
#school_data_complete.describe()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
count_schools = school_data["school_name"].count()
#count_schools
count_students = float(school_data_complete["Student ID"].count())
#count_students
total_budget = school_data["budget"].sum()
#total_budget
math_average = school_data_complete["math_score"].mean()
#math_average
average_reading_score = school_data_complete["reading_score"].mean()
#average_reading_score
overall_average_score = (math_average + average_reading_score)/2
#overall_average_score
passing_math_score = school_data_complete[(school_data_complete["math_score"]>=70)].count()["student_name"]
passing_reading_score = school_data_complete[(school_data_complete["reading_score"]>=70)].count()["student_name"]
passing_math_per = passing_math_score/count_students*100
passing_reading_per = passing_reading_score/count_students*100
passing_reading_per

district_summary = pd.DataFrame({"Total Schools":[count_schools], 
                                "Total Students":[count_students], 
                                "Total Budget":[total_budget], 
                                "Average Math Score":[math_average],
                                "Average reading Score": [average_reading_score],
                                "Overall Passing Rate": [overall_average_score], 
                                "Passing Math Score": [passing_math_per], 
                                "Passing Reading Score":[passing_reading_per]})
district_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
student_count = school_data_complete.groupby(["school_name"]).count()["Student ID"]
#student_count
school_data_2 = school_data.set_index(["school_name"])
#school_data_2
school_type = school_data_2["type"]
#school_type
school_budget = school_data_2["budget"]
#school_budget
student_budget = school_budget/student_count
#student_budget
average_math_score = school_data_complete.groupby(["school_name"]).mean()["math_score"]
#average_math_score
average_reading_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
#average_reading_score
#passing_math = school_data_complete[(school_data_complete["math_score"]>=70)].count()["student_name"]
passing_math = school_data_complete.loc[school_data_complete["math_score"]>=70].groupby(["school_name"]).size()
#passing_math
passing_reading = school_data_complete.loc[(school_data_complete["reading_score"]>=70)].groupby(["school_name"]).size()
#passing_reading
passing_math_per = passing_math/student_count*100
#passing_math_per
passing_reading_per = passing_reading/student_count*100
overall_passing_rate = (passing_math_per + passing_reading_per)/2
overall_passing_rate
```

```python
#Create a dataframe to hold the above results
school_summary = pd.DataFrame({"Total Students": student_count, "School Type": school_type, "School Budget": school_budget,
                                  "Budget per Student":student_budget, "Average Math Score": average_math_score,
                                 "Average Reading Score": average_reading_score, "Percentage Math": passing_math_per,
                                  "Percentage Reading": passing_reading_per, "Overall Passing Rate": overall_passing_rate})
school_summary.head()
```

```python
#Top Peforming Schools
top_performing_schools = school_summary.sort_values(["Overall Passing Rate"], ascending=False)
top_performing_schools.iloc[0:5,:]
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
school_summary_spending = school_summary.loc[:, ["Average Math Score", "Average Reading Score","Percentage Math", 
                   "Percentage Reading", "Overall Passing Rate", "Budget per Student"]]
#school_summary_spending
school_summary_spending["Average Spending Ranges"] = pd.cut(school_summary_spending["Budget per Student"], spending_bins, labels=group_names)
school_summary_spending
#school_summary_spending.groupby(["Average Spending Ranges"])
school_summary_spending.reset_index()
school_summary_spending.groupby(["Average Spending Ranges"]).mean()
```
